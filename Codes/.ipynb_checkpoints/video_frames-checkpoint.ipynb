```python
import cv2
import os
```

```python
try:
    if not os.path.exists('jhin'):
        os.makedirs('jhin')
except OSError:
    print ('Error: Creating directory of data')
```

```python
path = '../data/video_data/'
dest = '../data/image_data/'
```

```python
def capture_frames(path,dest):
    videos = [os.path.join(root, name)
             for root, dirs, files in os.walk(path)
             for name in files
              if name.endswith((".avi"))]
    currentframe=0         
    for vid in videos:
        cap = cv2.VideoCapture(vid)
        i = 0
        frame_skip = 30

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            if i > frame_skip - 1:
                name = dest + str(currentframe) + '.jpg'
                print ('Creating...' + name)
                cv2.imwrite(name, frame)
                currentframe += 1
                i = 0
                continue
            i += 1
        cap.release()
        cv2.destroyAllWindows()
```

```python
path1 = path+'/test/fire/'
dest1 = dest+'/test/fire/'
capture_frames(path1,dest1)
path2 = path+'/test/neutral/'
dest2 = dest+'/test/neutral/'
capture_frames(path2,dest2)
path3 = path+'/test/smoke/'
dest3 = dest+'/test/smoke/'
capture_frames(path3,dest3)
```

```python
path1 = path+'/train/fire/'
dest1 = dest+'/train/fire/'
capture_frames(path1,dest1)
path2 = path+'/train/neutral/'
dest2 = dest+'/train/neutral/'
capture_frames(path2,dest2)
path3 = path+'/train/smoke/'
dest3 = dest+'/train/smoke/'
capture_frames(path3,dest3)
```
